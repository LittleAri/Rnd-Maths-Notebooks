# Finding happy numbers and analysing their digit sums 

```python
def is_happy(n):
    digit_sums = []
    t = n
    while (t != 4 and t!=1):
        new_sum = get_digit_square_sums(t)
        digit_sums.append(new_sum)
        t = new_sum
    if t!=1:
        return "unhappy"
    else:
        return "happy"
    
```

```python
def get_digit_square_sums(n):
    new_x = 0
    for i in str(n):
        t = int(i)
        new_x += t**2
    return int(new_x)
```

```python
happy_numbers = []
for i in range(0,100000)[1:]:
    if is_happy(i) == 'happy':
        happy_numbers.append(i)
```

```python
def final_digit_sum(n):
    t = n
    while (len(str(t)) != 1):
        t = get_digit_sums(t)
    return t
```

```python
def get_digit_sums(n):
    new_x = 0
    for i in str(n):
        t = int(i)
        new_x += t
    return int(new_x)
```

```python
happy_number_digit_sums = []
for i in happy_numbers:
    happy_number_digit_sums.append(final_digit_sum(i))
```

```python
import matplotlib.pyplot as plt
import plotly.plotly as py
```

```python
fig = plt.figure(figsize=(15,4))
ax = fig.add_subplot(111)
ax.set_xlabel('number')
ax.set_ylabel('digit sum')
ax.plot(happy_numbers[:1000],happy_number_digit_sums[:1000], alpha = 0.2)
plt.ylim([0,10])
plt.xlim([0,round(max(happy_numbers[:1000]),-3)])
plt.show()
```
