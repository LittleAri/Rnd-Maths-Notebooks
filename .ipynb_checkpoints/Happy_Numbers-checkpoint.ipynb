# Finding happy numbers and analysing their digit sums 

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.plotly as py
import pandas as pd
```

```python
def get_digit_square_sums(n):
    new_x = 0
    for i in str(n):
        t = int(i)
        new_x += t**2
    return int(new_x)
```

```python
def get_happy_numbers_lessthan(x):
    happy_numbers = []
    unhappy_numbers = []
    
    for i in range (1,x):
        #print i
        sequence = []
        t = i
        while (t!=4 and t!=1 and t not in unhappy_numbers):
            t = get_digit_square_sums(t)
            sequence.append(t)
        if t==1:
            happy_numbers.append(i)
        else:
            for u in sequence:
                unhappy_numbers.append(u)
    return happy_numbers
                
```

```python
def final_digit_sum(n):
    t = n
    while (len(str(t)) != 1):
        t = get_digit_sums(t)
    return t
```

```python
def get_digit_sums(n):
    new_x = 0
    for i in str(n):
        t = int(i)
        new_x += t
    return int(new_x)
```

```python
u = 10000
happy_numbers = get_happy_numbers_lessthan(u)
digit_sums_of_happy_numbers = []
for i in range(0,u):
    if i in happy_numbers:
        digit_sums_of_happy_numbers.append(final_digit_sum(i))
        
        
new_str = str(digit_sums_of_happy_numbers)
new_str2 = new_str.replace(',','').replace(' ','').replace('[','').replace(']','')
df = pd.DataFrame({'Count':list(new_str2)})
print "--- Digit Sums for Happy Numbers under", u,"---"
print "\n"
table = df.Count.value_counts()
print table
print "\n"
plt.clf()
fig = plt.figure(figsize=(15,7))
ax = fig.add_subplot(111)
ax.set_title('Digit Sums of Happy Numbers and the Average Digit Sum')
ax.set_xlabel('Number')
ax.set_ylabel('Digit Sum')
ax.plot(happy_numbers,digit_sums_of_happy_numbers, alpha = 0.4)
ax.plot(happy_numbers, [np.mean(digit_sums_of_happy_numbers[:happy_numbers.index(i)]) for i in happy_numbers],"m")
plt.ylim([0,10])


plt.show()
```
