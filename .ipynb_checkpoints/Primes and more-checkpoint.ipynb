```python
import matplotlib.pyplot as plt
%matplotlib inline



 
# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
#print "Current size:", fig_size
 
# Set figure width to 12 and height to 9
fig_size[0] = 12
fig_size[1] = 5
plt.rcParams["figure.figsize"] = fig_size
#print "New size", fig_size
primes = []

r = 10000
for n in range(1,r):
    z = range(1,n-1)
    a = 0
   
    for i in z:
        if n%i==0:
            a+=1
        else:
            pass
    if a==0 or a==1:
        primes.append(n)
        
prime_or_not = []
for t in range(1,r):
    if t in primes:
        prime_or_not.append(1)
    else:
        prime_or_not.append(0)
        
my_list = range(1,r)
plt.plot(my_list,prime_or_not, 'c')
    
triangular_numbers = []
for n in range(1,r):
    if n*(n+1)/2 < r:
        triangular_numbers.append(n*(n+1)/2)
    else:
        break


triangular_or_not = []
for t in range(1,r):
    if t in triangular_numbers:
        triangular_or_not.append(1)
    else:
        triangular_or_not.append(0)
        
plt.plot(my_list,triangular_or_not, 'b')



perfect_numbers = []
for n in range(1,r):
    z = range(1,n-1)
    b = []
    if n not in primes:
   
        for i in z:
            if n%i==0:
                b.append(i)
            else:
                pass
        if sum(b)==n:
            perfect_numbers.append(n)
        
perfect_or_not = []
for t in range(1,r):
    if t in perfect_numbers:
        perfect_or_not.append(1)
    else:
        perfect_or_not.append(0)
        
plt.plot(my_list,perfect_or_not, 'm')


        

cur_axes = plt.gca()
cur_axes.axes.get_yaxis().set_visible(False)

cur_axes.set_title("A colourful representation of beautiful numbers.")
    
print "Prime numbers <", r
print len(primes)

print "\nTriangular numbers <", r
print len(triangular_numbers)

print "\nPerfect numbers <", r
print len(perfect_numbers)
    
```
